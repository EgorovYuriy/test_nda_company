Создадим новую колонку с названием района. Некоторые районы повторяются с отличием от стороны берега, поиск по совпадению по районам будем делать в зависимости от длины названия - более специфичные названия будут стоять в первую очередь

```python
districts = [
    'орджоникидзевский (левый берег)', 
    'ленинский (правый берег)',
    'ленинский (левый берег)',
    'орджоникидзевский', 
    'ленинский', 
    'правобережный'
]

def find_district(full_address):
    for title in districts:
        if title in full_address:
            return title
    return ''

df['district'] = df.full_address.apply(find_district)
df['address'] = df.full_address.apply(lambda x: x.replace(find_district(x), '').strip())
df.district = df.district.replace('', np.nan)
df.drop('full_address', axis = 1, inplace = True)

df.sample(10)
```

### 2.6 Колонка District

заполним пропуски в столбце район методом машинного обучения: создадим новый датасет без пропусков, предсказанные значения подставим на места пропусков в исходном датасете

```python
df_not_na = df.dropna(subset=['district'])

encoder_street = LabelEncoder()
encoder_house_number = LabelEncoder()

for d in (df, df_not_na):
    
    d['street_encoded'] = encoder_street.fit_transform(d.street)
    d['house_number_encoded'] = encoder_house_number.fit_transform(d.house_number)

X, y = df_not_na[['street_encoded', 'house_number_encoded']], df_not_na.district

model = RandomForestClassifier(random_state=RS)
model.fit(X, y)

df.sample(3)
```

закодированные колонки добавлены, теперь сделаем по ним предсказания и заполним пропуски

```python
X_na = df[df.district.isna()][['street_encoded', 'house_number_encoded']]
pred_districts = model.predict(X_na)
df.loc[df.district.isna(), 'district'] = pred_districts

df.drop(['street_encoded', 'house_number_encoded'], axis=1, inplace=True)
del df_not_na
```

### 3.1 Подготовка признаков

поменяем тип категориальных колонок для обучения бустинга

```python
def df_change_type_columns(data):
    
    for i in data.select_dtypes('object').columns.to_list():
        data[i] = data[i].astype('category')
    
    return df

df = df_change_type_columns(df)
df.info()
```

подготовим пайплайн для обучения моделей с учетом их особенностей для препроцессинга

```python
X, y = df.drop(['price'], axis = 1), df.price

# категориальные признаки, где количество уникальных значений меньше или равно 5
cat_ohe = ['district']

# остальные категориальные признаки
cat_oe = ['street', 'house_number']

cat_num = X.select_dtypes(exclude = 'category').columns.to_list()

preprocessor_lin = ColumnTransformer(
                 [('catigorical_ohe', OneHotEncoder(handle_unknown = 'ignore', drop = 'first', sparse = False), cat_ohe),
                  ('catigorical_oe', OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1), cat_oe),
                  ('scaler', StandardScaler(), cat_num)],
                    remainder = 'passthrough'
)

preprocessor_trees = ColumnTransformer(
                 [('catigorical_ohe', OneHotEncoder(handle_unknown = 'ignore', drop = 'first', sparse = False), cat_ohe),
                  ('catigorical_oe', OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1), cat_oe)],
                    remainder = 'passthrough'
)
```

пропишем функцию для обучения моделей

```python
def regressor_cv(model, params):
    
    regressor = RandomizedSearchCV(
         estimator = model, 
         param_distributions = params, 
         cv = KFold(5, shuffle = True, random_state = RS),
         scoring = 'neg_mean_absolute_error',
         n_jobs = -1,
         verbose = 10
    )
    regressor.fit(X, y)
    
    score_cv = -(regressor.best_score_).round(2)
    best_params = regressor.best_params_
    best_model = regressor.best_estimator_
    
    print('-'*23)
    print('Simulated successfull')
    print('-'*23)
    return score_cv, best_params, best_model
```

### 3.2 Обучение моделей

#### 3.2.1 Linear Regression

```python
%%time

pipe_lr = Pipeline([
    ('preprocessor_lin', preprocessor_lin),
    ('lr', LinearRegression())
])

params_lr = {}

lr_score__cv, lr_best_params, lr_best_model = regressor_cv(
    pipe_lr, 
    params_lr
)
print(lr_score__cv)
print(lr_best_params)
```

#### 3.2.2 Random Forest

```python
%%time

pipe_rf = Pipeline([
    ('preprocessor_trees', preprocessor_trees),
    ('rf', RandomForestRegressor(random_state = RS))
])
params_rf = {
    'rf__n_estimators':[100,500],
    'rf__max_depth':[5,10,20]
}
rf_score_cv, rf_best_params, rf_best_model = regressor_cv(
    pipe_rf, 
    params_rf
)
print(rf_score_cv)
print(rf_best_params)
```

#### 3.2.3 LightGBM

```python
%%time

pipe_lgbm = Pipeline([
     ('lgbm', LGBMRegressor(random_state = RS))
])

params_lgbm = {
    'lgbm__n_estimators':[50,100]
}

lgbm_score_cv, lgbm_best_params, lgbm_best_model = regressor_cv(
    pipe_lgbm, 
    params_lgbm
)
print(lgbm_score_cv)
print(lgbm_best_params)
```

лучший скор показал случайный лес. соберем заново модель на всех данных для сохранения и выкатывания в прод.

## 4. Оформление и тестрирование REST API (внимание, ячейка закомментирована, инструкция внутри)

функция на тест запроса flask-приложения, для тестировани нужно раскомментировать

```python
def json_request(rooms: int,
                 floor: int,
                 total_floor: int,
                 total_area: float,
                 living_area: float,
                 kitchen_area: float,
                 district: str,
                 street: str,
                 house_number: str):

    url = "http://127.0.0.1:5000/predict"

    data = {
        "rooms": rooms,
        "floor": floor,
        "total_floor": total_floor,
        "total_area": total_area,
        "living_area": living_area,
        "kitchen_area": kitchen_area,
        "district": district,
        "street": street,
        "house_number": house_number
    }

    response = requests.post(url, json=data)

    print(response.json())
```

по требованию заказчика нужно вывести предсказания для 3х объектов. для подтверждения корректности работы веб-приложения и сравнения предсказаний с реальными значениями сначала запускаем ячеку с 3мя выборочными объектами, затем заносим параметры выбранных объектов в функции и запускаем по очереди, под каждой будет выведен результат. в каждой ячейке ка кобразец введены показательные данные

### 5.1 Kmeans

теперь проведем кластеризацию данных по выбранным колонкам. сначала проверим методом локтя

```python
X_kmeans = df[
    ['price',
     'rooms', 
     'total_area', 
     'living_area',
     'kitchen_area']
]

# Нормализация данных
scaler = StandardScaler()
X_kmeans_scalered = scaler.fit_transform(X_kmeans)

# Определение оптимального количества кластеров (метод локтя)
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=RS)
    kmeans.fit(X_kmeans_scalered)
    inertia.append(kmeans.inertia_)

# Визуализация методом локтя
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('номер кластера')
plt.ylabel('Inertia')
plt.title('Кластеризация - метод локтя')
plt.grid(True)
plt.show()
```

Метод Kmeans

значением количества кластеров выбираем число 4, т.к. на этой отметке согласно графику значение инерции стабилизируется

создадим новые столбцы с номером кластера и предсказанным районом согласно номеру кластера

```python
# разбивка на 5 кластеров в соответствие с требованием заказчика
kmeans = KMeans(n_clusters=4, random_state=42)

# новый столбец с результатом кластеризации
df['district_cluster_kmeans'] = kmeans.fit_predict(X_kmeans)

# словарь, который сопоставляет номера кластеров и районы
cluster_to_district = {}

for cluster in df.district_cluster_kmeans.unique():
    
    district_value = df[df.district_cluster_kmeans == cluster]['district'].mode().values[0]
    cluster_to_district[cluster] = district_value

# новый столбец с результатом предсказания кластера
df['district_predict_kmeans'] = df.district_cluster_kmeans.map(cluster_to_district)
df.sample(3)
```

теперь выберем все объекты, в которых предсказанное значение района по кластеру совпало с действительным значением и посчитаем процент точности

```python
# количество строк, в которых значения 'district' и 'district_predict' совпадают
matching_rows = df[df.district == df.district_predict_kmeans]

# процент совпадения
percentage_matching = len(matching_rows) / len(df) * 100

print(f"Cовпадения между районами и предсказаниями методом Kmeans = {percentage_matching:.0f}%")
```

Разбить данные на кластера так, чтобы каждому кластеру соответствовало только одно уникальное значение столбца 'district' - невыполнимая задача. Проблема заключается в том, что алгоритмы кластеризации стремятся группировать данные на основе схожих признаков, и они не знают о наличии определенных уникальных значений в столбце 'district', который не был определен в список выбранных для кластеризации признаков. Кластеризация, как правило, используется для выявления общих паттернов в данных и различных группировок для снижения дисперсии, а не для точного соответствия уникальных значений.

### 5.2 DBSCAN

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN

eps_values = np.linspace(0.1, 2.0, 20)
min_samples_values = [3, 5, 10, 15]

# функция для кластеризации с DBSCAN и сопоставления с районами
def cluster_and_map_districts(X, eps_values, min_samples_values):
    
    best_score = -1
    best_eps = None
    best_min_samples = None
    best_labels = None

    for eps in eps_values:
        for min_samples in min_samples_values:
            
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels_dbscan = dbscan.fit_predict(X)
            
            if len(set(labels_dbscan)) > 1:
                score = silhouette_score(X, labels_dbscan)
                if score > best_score:
                    
                    best_score = score
                    best_eps = eps
                    best_min_samples = min_samples
                    best_labels = labels_dbscan

    df['district_cluster_dbscan'] = labels_dbscan

    # словарь для сопоставления кластеров с районами
    cluster_to_district = {}
    for cluster in df.district_cluster_dbscan.unique():
        
        mode_district = df[df.district_cluster_dbscan == cluster]['district'].mode().values[0]
        cluster_to_district[cluster] = mode_district

    # новый столбец для предсказанных районов на основе кластеров
    df['district_predict_dbscan'] = df.district_cluster_dbscan.map(cluster_to_district)

    return best_eps, best_min_samples, labels_dbscan

best_eps, best_min_samples, labels_dbscan = cluster_and_map_districts(X_kmeans_scalered, eps_values, min_samples_values)

print("Best EPS:", best_eps)
print("Best Min Samples:", best_min_samples)
```

```python
df['district_cluster_dbscan'] = labels_dbscan
df.sample(5)
```

```python
# количество строк, в которых значения 'district' и 'district_predict' совпадают
matching_rows = df[df.district == df.district_predict_dbscan]

# процент совпадения
percentage_matching = len(matching_rows) / len(df) * 100

print(f"Процент совпадения между районами и предсказаниями методом DBSCAN = {percentage_matching:.0f}%")
```

### 5.3 PCA

PCA KMEANS

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_kmeans_scalered)

unique_labels = df.district_cluster_kmeans.unique()
num_clusters = len(unique_labels)

# Создаем цветовую карту, которая будет содержать уникальный цвет для каждого кластера
colors = plt.cm.viridis(np.linspace(0, 1, num_clusters))

# Визуализируем данные и создаем легенду
plt.figure(figsize=(10, 6))
for i, label in enumerate(unique_labels):
    cluster_data = X_pca[df.district_cluster_kmeans == label]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], c=colors[i], label=f'Cluster {label}', s=50)

plt.title('DBSCAN Clustering (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

plt.show()
```
